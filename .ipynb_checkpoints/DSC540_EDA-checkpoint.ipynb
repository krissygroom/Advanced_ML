## Introduction

> My dataset of interest was found at https://data.world/carolee/women-in-movies with the original data found at:  https://github.com/fivethirtyeight/data/tree/master/bechdel.  The dataset references a 2014 article from FiveThirtyEight.com – found at: https://fivethirtyeight.com/features/the-dollar-and-cents-case-against-hollywoods-exclusion-of-women/ and is an article that analyzes data about movies that either pass or do not pass the Bechdel Test.  The Bechdel Test is a test that measures Hollywood bias in that it specifies that “if a movie can satisfy three criteria — there are at least two named women in the picture, they have a conversation with each other at some point, and that conversation isn’t about a male character — then it passes “The Rule,” whereby female characters are allocated a bare minimum of depth”(Hickey, 2014).  
The dataset consists of films from 1970 to 2013, with 15 features and 1796 instances with approximately 200 missing values.  The features include label features, categorical features with PASS/NO PASS as values as well as more specific categories of passing one or more of the criteria, year and period categories, and also numeric features such as budget and revenue.  I will be considering international revenue as my target variable of interest and my question is: Do movies that pass the Bechdel test differ in their international success as measured by movie revenue from movies that do not pass the test.

### Data Cleaning:
> In this step of the analysis, I will transform any data in order to update the dataset into a more
> consistent and uniform dataset.

I see that there are missing values particularly with period code and decade code, although there are some recent movies with missing domgross and intgross values.  I can easily fill in the values for the period code and decade code but I will remove the observations with missing values of domgross and intgross so as to avoid changing the data very much.

```python
# Add period code and decade code to observations with missing values which began earlier than 1989
# 1989 - 1985
df.loc[((df.year <= 1989) & (df.year > 1984)), 'period code'] = 6
df.loc[((df.year <= 1989) & (df.year > 1984)), 'decade code'] = 4

# 1984 - 1980
df.loc[((df.year <= 1984) & (df.year >= 1980)), 'period code'] = 7
df.loc[((df.year <= 1984) & (df.year >= 1980)), 'decade code'] = 4

# 1979 - 1975
df.loc[((df.year <= 1979) & (df.year >= 1975)), 'period code'] = 8
df.loc[((df.year <= 1979) & (df.year >= 1975)), 'decade code'] = 5

# 1974 - 1970
df.loc[((df.year <= 1974) & (df.year >= 1970)), 'period code'] = 9
df.loc[((df.year <= 1974) & (df.year >= 1970)), 'decade code'] = 5
```

However, I would like to explore a bit further.

```python
# visualize the correlations
# using template found at: https://www.kaggle.com/chapagain/house-prices-eda-data-modelling-and-prediction

import seaborn as sns
sns.set(style="whitegrid", color_codes=True)

corr_matrix = df_table.corr()

sns.set(font_scale=1.10)
plt.figure(figsize=(15, 15))

sns.heatmap(corr_matrix, vmax=.8, linewidths=0.01,
            square=True, annot=True, cmap='viridis', linecolor="white")

plt.title('Correlation between features');
```

```python
# averages of budget and revenues according to whether the movie passed or failed
budget_mean = df_table.groupby('binary_FAIL')['budget'].mean()
```

```python
d_gross_mean = df_table.groupby('binary_FAIL')['domgross'].mean()
```

```python
int_gross_mean = df_table.groupby('binary_FAIL')['intgross'].mean()
```

```python
# looking at median to account for outliers
budget_med = df_table.groupby('binary_FAIL')['budget'].median()
d_gross_med = df_table.groupby('binary_FAIL')['domgross'].median()
int_gross_med = df_table.groupby('binary_FAIL')['intgross'].median()
```

```python
# visualize budget
plt.figure(figsize=(10,8))
plt.bar(budget_mean.index, budget_mean.values);

plt.xticks([0, 1], labels=['PASS', 'FAIL'])
plt.title('Average Budgets for Movies Passing and Failing the Bechdel Test')
plt.xlabel('Pass or Fail Bechdel Test')
plt.ylabel('Average Budget');
```

```python
# visualize gross revenue
plt.figure(figsize=(10,8))
plt.bar(d_gross_mean.index, d_gross_mean.values);

plt.xticks([0, 1], labels=['PASS', 'FAIL'])
plt.title('Average Domestic Revenue for Movies Passing and Failing the Bechdel Test')
plt.xlabel('Pass or Fail Bechdel Test')
plt.ylabel('Average Domestic Revenue');
```

```python
# visualize international revenue
plt.figure(figsize=(10,8))
plt.bar(int_gross_mean.index, int_gross_mean.values);

plt.xticks([0, 1], labels=['PASS', 'FAIL'])
plt.title('Average International Revenue for Movies Passing and Failing the Bechdel Test')
plt.xlabel('Pass or Fail Bechdel Test')
plt.ylabel('Average International Revenue');
```

```python
# visualize median budget
plt.figure(figsize=(10,8))
plt.bar(budget_med.index, budget_med.values);

plt.xticks([0, 1], labels=['PASS', 'FAIL'])
plt.title('Median Budgets for Movies Passing and Failing the Bechdel Test')
plt.xlabel('Pass or Fail Bechdel Test')
plt.ylabel('Median Budget');
```

```python
# visualize median gross revenue
plt.figure(figsize=(10,8))
plt.bar(d_gross_med.index, d_gross_med.values);

plt.xticks([0, 1], labels=['PASS', 'FAIL'])
plt.title('Median Domestic Revenue for Movies Passing and Failing the Bechdel Test')
plt.xlabel('Pass or Fail Bechdel Test')
plt.ylabel('Median Domestic Revenue');
```

```python
# visualize international revenue
plt.figure(figsize=(10,8))
plt.bar(int_gross_med.index, int_gross_med.values);

plt.xticks([0, 1], labels=['PASS', 'FAIL'])
plt.title('Median International Revenue for Movies Passing and Failing the Bechdel Test')
plt.xlabel('Pass or Fail Bechdel Test')
plt.ylabel('Median International Revenue');
```

From these basic visualizations, there appears to be differences between the budget allotted passing vs failing movies, and a smaller difference of domestic revenue for passing and failing movies.  International revenue appears to have a much smaller difference.  However, I did not check confidence intervals to see if the differences are significant, so I cannot say with confidence if there is a significant difference between the budgets, domestic revenue, and international revenue of movies that pass and movies that fail the Bechdel test.  I also did not account for budget being strongly correlated with revenue as was performed in the original analysis.  Seeing that movies passing the Bechdel test receive less budget, it would make sense that the revenue would generally be less.
